```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('animals_weight.txt', delim_whitespace=True)

df
```

```python
df.isnull().sum()
```

checking if anyones brain is heavier than their whole body!

```python
df = df.drop(df.index[df.apply(lambda row: row['Brain'] >= row['Body'], axis=1)])

df.reset_index(inplace=True, drop=True)

df
```

looks like there were some :)

```python
X = df.drop(columns=["Body"])
Y = df["Body"]
```

```python
scaler = StandardScaler()
Scaled_X = scaler.fit_transform(X)
```

Split by 20 percent:

```python
X_train, X_test, Y_train, Y_test = train_test_split(Scaled_X, Y, test_size=0.2, random_state=42)
```

```python
KNN_1 = KNeighborsRegressor(n_neighbors=1)
KNN_3 = KNeighborsRegressor(n_neighbors=3)
KNN_5 = KNeighborsRegressor(n_neighbors=5)
KNN_1.fit(X_train, Y_train)
KNN_3.fit(X_train, Y_train)
KNN_5.fit(X_train, Y_train)
```

now we predict weight

```python
input = [53.298, 1247.122, 0.583, 0.583, 0.041]

predictions = {
    "KNN_1": [],
    "KNN_3": [],
    "KNN_5": []
}

for value in input:
    scaled_value = scaler.transform(np.array([value]).reshape(1, -1))
    
    predictions["KNN_1"].append(KNN_1.predict(scaled_value)[0])
    predictions["KNN_3"].append(KNN_3.predict(scaled_value)[0])
    predictions["KNN_5"].append(KNN_5.predict(scaled_value)[0])

No1_1 = predictions["KNN_1"]
No3_1 = predictions["KNN_3"]
No5_1 = predictions["KNN_5"]

print(np.float64(No1_1))
print(np.float64(No3_1))
print(np.float64(No5_1))
```
